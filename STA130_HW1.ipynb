## **6.**

`count`: Provides the number of non-missing values for each column.

`mean`: The average value of all non-missing values for each column.

`std`: The standard deviation of all non-missing values for each column. The higher the standard deviation, the more the data spread in a wider range

`min`: The lowest value of all non-missing values for each column.

`25%`: The value such that exactly 25% of all non-missing values are lower than the value itself, for each column.

`50%`: The median value of all non-missing values for each column. Median value is a value such that the number of values below the median equals the number of values above the median.

`75%`: The value such that exactly 75% of all non-missing values are lower than the value itself, for each column

`max`: The highest value of all non-missing values for each column.

## **7.**

1. When you want to remove rows (or columns if you use `df.dropna(axis=1)`) that contains missing values.
2. When you want to remove columns that is considered as not useful for the dataset (e.g. irrelevant columns, columns with too many missing data)
3. Applying `del df['col']` before `df.dropna()` could be important in the case that you want to remove rows and columns that are not useful (as mentioned in 2.). Since executing `del df['col']` removes columns with too many missing data, doing that before `df.dropna()` preserves some rows (or observations) that would otherwise be deleted by `df.dropna()` since they only have the missing data in the "not useful" column.

```python
df2.isna().sum()
```

```python
df2_dropdel = df2.dropna()
del df2_dropdel['deck'], df2_dropdel['age']
df2_dropdel.describe(include='all')
```

```python
df2_deldrop = df2.copy()
del df2_deldrop['deck'], df2_deldrop['age']
df2_deldrop = df2_deldrop.dropna()
df2_deldrop.describe(include='all')
```

4. In the example above, I use the dataset that contains a lot of missing data on `age` and `deck`. Here, if I use `df.dropna()` before `del df['col']` (in this case the `col` are `age` and `deck`), there are only 182 data left. However, if I use `del df['col']` before `df.dropna()`, there are still 889 data left (only 2 short from the original due to 2 missing values on the `embarked` column). So, if I think that a column has too much missing values, it's better to do `del df['col']` before `df.dropna()` rather than the reverse (or just `df.dropna()`).

1. `df.groupby("col1")["col2"].describe()` outputs a table that is fundamentally provides the same data as of what `df.describe()` would do (like mean, min, max, etc.), but instead of counting it for the whole dataset, it divides the dataset into groups first, according to the number of unique data in `col1` (for example, if there are 4 unique values, then there are 4 groups). Then, after that the code will provide the summarization of each group in the form of a table. Demonstration is shown below.

```python
df2.groupby("pclass")["age"].describe()
```

```python
df2.describe()
```

2. `count` variable in `df.groupby("col1")["col2"].describe()` only counts the number of non-missing data in the column `col2`, in respect to their `col1` grouping, while `df.describe()` counts number of non-missing data for each column in the dataset.
